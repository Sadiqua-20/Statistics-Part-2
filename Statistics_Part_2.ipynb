## Q1.  What is hypothesis testing in statistics?

ANS 1.

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses and using statistical techniques to determine which one is more likely to be true.

##Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?


ANS 2.
A **null hypothesis** **($H_{0}$)** is a statement that assumes no effect, no difference, or no relationship in the population. It serves as the default assumption that we aim to test.

An **alternative hypothesis ($𝐻_{𝐴}$ or $𝐻_{1}$)** contradicts the null hypothesis, suggesting that there is an effect, a difference, or a relationship in the data.

- **Example (Mean Test):**  
$H_{0}$ : "The average height of students is 170 cm."  
$H_{𝐴}$ : "The average height of students is not 170 cm."

- **Difference between $𝐻_{𝐴}$ and $𝐻_{1}$:**  
 1. $H_{0}$ is the default hypothesis while $H_{𝐴}$ always contradicts $H_{0}$.
 2. $H_{0}$ always has $=$ sign while $H_{𝐴}$ has inequality($\neq$, >, <).


##Q3. What is the significance level in hypothesis testing, and why is it important?

ANS 3.

The **significance level ($\alpha$)** is the probability of rejecting the null hypothesis ($H_{0}$) when it is actually true. It represents the risk of making a Type I error (false positive).

**Importance of significance level ($\alpha$):**
- **Controls Error Rate:** Limits the probability of mistakenly concluding that an effect exists when it actually doesn’t.
- **Determines Decision Threshold:** Helps decide whether to reject $H_{0}$ based on the p-value.
- B**alances Risk:** A lower significance level ($\alpha$) reduces false positives but increases the risk of a false negative (Type II error).

##Q4. What does a P-value represent in hypothesis testing?

ANS 4.

The **P-value (probability value)** is a measure of the strength of evidence against the null hypothesis ($H_{0}$). It represents the probability of obtaining a test result as extreme than the observed data, assuming that $H_{0}$ is true.   

In other words, the P-value is the probability of obtaining the observed result (or a more extreme result) by chance, if there is no real effect or relationship.

P-values are often misinterpreted as the probability of the null hypothesis being true. However, P-values only indicate the probability of observing the data (or more extreme data) assuming the null hypothesis is true.

##Q5. How do you interpret the P-value in hypothesis testing?

ANS 5.

1. **Range:** P-values range from 0 to 1, where 0 indicates that the observed result is impossible under the null hypothesis, and 1 indicates that the observed result is perfectly consistent with the null hypothesis.

2. **Interpretation:** A small P-value (typically ≤ 0.05) indicates that the observed result is unlikely to occur by chance, and therefore, the null hypothesis can be rejected. A large P-value indicates that the observed result is likely to occur by chance, and therefore, the null hypothesis cannot be rejected.

##Q6. What are Type 1 and Type 2 errors in hypothesis testing?

ANS 6. In hypothesis testing, two types of errors can occur when making a decision about the null hypothesis ($H_{0}$):  

1. **Type I Error (False Positive):**  
   - Occurs when $H_{0}$ is true, but we mistakenly reject it.  
   - Represents a *false alarm* — detecting an effect when there isn't one.  
   - The probability of making a Type I error is denoted by $\alpha$(significance level).  
   - **Example:** A medical test wrongly detects a disease in a healthy person.  

2. **Type II Error (False Negative):**  
   - Occurs when $H_{0}$ is false, but we fail to reject it.  
   - Represents a *missed detection*—failing to recognize an actual effect.  
   - The probability of making a Type II error is denoted by $\beta$.  
   - **Example:** A faulty medical test fails to detect a disease in a sick person.

##Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

ANS 7.

A **one-tailed test** and a **two-tailed test** are types of statistical tests used to determine whether to reject the null hypothesis ($H_{0}$) based on sample data. The key difference lies in how they test the alternative hypothesis ($H_{A}$).

**1. One-Tailed Test:**  
- Tests for an effect in only one direction (either greater or less than).  
- Alternative Hypothesis ($H_{A}$)  
 Example:  
  - Right-tailed test: $\( H_A: \mu > \mu_0 \$) (greater than)  

  - Left-tailed test: $\( H_A: \mu < \mu_0 \$) (less than)  
- Used when the research question expects a specific directional outcome.  
- **Example:** A company claims a new drug increases heart rate. The test only checks if the heart rate increases (not decreases).  

**2. Two-Tailed Test:**  
- Tests for an effect in both directions(greater or less than).  
- Alternative Hypothesis ($H_{A}$)   
 Example:  
  - $H_{A}$ : $\mu$ $\neq$ $\mu_0$
- Used when the research question is open-ended and does not specify direction.  
- **Example:** Testing if a new teaching method affects test scores (it could increase **or** decrease scores).

##Q8. What is the Z-test, and when is it used in hypothesis testing?

ANS 8.

A Z-test is a statistical test used to determine whether there is a significant difference between sample and population parameters (like means or proportions) when the population variance(or population standard deviation) is known and the sample size is large ($n$ > 30).  

**When to Use a Z-Test?**   
A Z-test is appropriate when:  
1. **Sample size is large ($n$ > 30) :** Central Limit Theorem ensures normality.  
2. **Population variance ($\sigma^2$) or population standard deviation ($\sigma$) is known:** Unlike the T-test, which is used when variance is unknown.  
3. **Data follows a normal distribution :** Especially for smaller samples.  


**Formula for One-Sample Z-Test :**  

$Z$ = $\frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$ ,
  
where:  
-  $\bar{x}$  = Sample mean  
-  $\mu$  = Population mean  
-  $\sigma$  = Population standard deviation  
-  $n$ = Sample size

## Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

ANS 9.

A **Z-score** (or standard score) measures how many standard deviations a data point or sample mean is from the population mean. It helps determine whether an observed result is significantly different from what is expected under the null hypothesis ($H_0$).  


**Calculation of Z-Score?**  

1. For an Individual Data Point:

$Z = \frac{X - \mu}{\sigma}$  

where,   
- $X$ = Individual data value  
- $\mu$ = Population mean  
- $\sigma$ = Population standard deviation  

2. For a Sample Mean (Z-Test Formula):  

$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$

where,  
- $\bar{x}$  = Sample mean  
- $\mu$  = Population mean  
- $\sigma$ = Population standard deviation  
- $n$  = Sample size  


**What Does the Z-Score Represent? :**  
Used to Determine Statistical Significance in Hypothesis Testing:  
   - Compare the Z-score with the critical value from the **Z-table** for a given significance level $\alpha$.  
   - If the Z-score is beyond the critical value, reject $H_0$ (suggesting a significant difference).

## Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

ANS 10.

**T-distribution:**  
- The T-distribution (also called **Student's t-distribution**) is a probability distribution used in hypothesis testing when the sample size is small and/or the population standard deviation ($\sigma$) is unknown.

- It is similar to the normal distribution but has heavier tails, meaning it accounts for more variability in small samples.  
- **Depends on Degrees of Freedom ($df = n - 1$) :**  
   - Smaller samples - Heavier tails (more variability).  
   - As $n$ increases, the t-distribution approaches the normal distribution.

**When to Use the T-Distribution Instead of the Normal Distribution :**  
1. **Sample Size is Small ($n < 30$)**
   The Central Limit Theorem states that larger samples ($n > 30$) tend to follow a normal distribution, but for small samples, the t-distribution is more accurate.  

2. **Population Standard Deviation ($\sigma$) is Unknown**  
   If $\sigma$ is unknown, we estimate it using the sample standard deviation ($s$), which adds uncertainty. The t-distribution accounts for this extra variability.  

3. **Data is Approximately Normally Distributed**  
   The t-distribution is valid if the sample data is approximately normal (especially for very small samples).

##Q11. What is the difference between a Z-test and a T-test?

ANS 11.

A **Z-test** and a **T-test** are both statistical tests used in hypothesis testing, but they are used under different conditions. The key difference lies in the *sample size* and *whether the population standard deviation ($\sigma$) is known*.  

**1. Z-Test :**  
- Used when:  
  - The sample size is large ($n \geq 30$).  
  - The *population standard deviation ($\sigma$) is known*.  
  - The data follows a normal distribution or $n$ is large (Central Limit Theorem applies).  

- **Formula for One-Sample Z-Test:**  
$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$   
where,  
  - $\bar{x}$ = Sample mean  
  - $\mu$ = Population mean  
  - $\sigma$ = Population standard deviation  
  - $n$ = Sample size  

- **Example :**  Testing if the average IQ of students is different from 100, given that the population standard deviation is known.  



**2. T-Test :**  
- Used when:  
  - The sample size is *small* ($n < 30$).  
  - The population standard deviation ($\sigma$) is unknown (use sample Standard Deviation ($s$) instead).  
  - The data is approximately normally distributed.  

- **Formula for One-Sample T-Test :**  
$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$  
where:  
  - $s$ = Sample standard deviation (since $\sigma$ is unknown)  
  - $\bar{x}$ = Sample mean  
  - $\mu$ = Population mean   
  - $n$ = Sample size

- **Example :**  Testing if a new diet affects body weight, based on a sample of 15 people, with no knowledge of the population standard deviation.

##Q12.What is the T-test, and how is it used in hypothesis testing?

ANS 12.

A **T-test** is a statistical test used to compare the means of a sample and a population when the population standard deviation ($\sigma$) is unknown. It determines whether the difference between means is statistically significant.  

**Steps in Hypothesis Testing Using a T-Test :**  
1. State Hypotheses :  
   - Null Hypothesis ($H_0$): No significant difference $\mu_1 = \mu_2$.  
   - Alternative Hypothesis ($H_A$) : There is a significant difference $\mu_1 \neq \mu_2$\.  

2. Choose Significance Level ($\alpha$) :  Common values: 0.05 (5%) or 0.01 (1%).  

3. Calculate the $t_{Statistics}$ using the formula:  
$t_{Statistics} = \frac{\bar{x} - \mu}{s / \sqrt{n}}$  
where:  
  - $s$ = Sample standard deviation (since $\sigma$ is unknown)  
  - $\bar{x}$ = Sample mean  
  - $\mu$ = Population mean   
  - $n$ = Sample size   

4. Find the Critical Value $(t_{Critical})$ from the t-table(based on degrees of freedom: $df = n - 1$).  

5. Compare $t_{Statistics}$ with $t_{Critical}$:  
   - If $t_{Statistics}$ $>$ $t_{Critical}$, reject $H_0$ (significant difference).  
   - If $t_{Statistics}$ $<$ $t_{Critical}$, fail to reject $H_0$ (no significant difference).

##Q14. What is a confidence interval, and how is it used to interpret statistical results?

ANS 14.

A **confidence interval (CI)** is a range of values used to estimate an unknown population parameter (such as a mean or proportion) based on a sample. It provides a range within which we expect the true population value to lie *with a certain level of confidence* (e.g., 95%).  


**How to Calculate a Confidence Interval :**  
For a population mean ($\mu$) with a known or estimated standard deviation:  

$CI = \bar{x} \pm (Z \text{ or } t) \times \frac{\sigma}{\sqrt{n}}$

where,  
- $\bar{x}$ = Sample mean  
- $Z$ or $t$ = Critical value (from the *Z-table* for large samples or *t-table* for small samples)  
- $\sigma$ = Population standard deviation (or sample SD $s$ if $\sigma$ is unknown)  
- $n$= Sample size

**How Confidence Intervals Help in Interpretation :**  
1. **Assessing Statistical Significance**  
   - If a confidence interval does not include a hypothesized value (e.g., population mean), we may reject the null hypothesis.  
   - Example: If a CI for the mean weight of a drug-treated group is *(65, 75) kg* and the normal weight is *60 kg*, we conclude the drug likely increases weight.  

2. **Margin of Error :**
   - The *wider* the confidence interval, the *more uncertainty* in the estimate.  
   - The *narrower* the interval, the *more precise* the estimate.  

3. **Comparing Groups :**  
   - If two groups have overlapping confidence intervals, their means might not be significantly different.  
   - If the intervals *do not overlap*, there is stronger evidence of a difference.  

##Q15. What is the margin of error, and how does it affect the confidence interval?

ANS 15.

The **margin of error (MoE)** is the amount by which the sample estimate (e.g., sample mean) could vary from the true population value. It reflects the uncertainty in the estimate and is used to construct a **confidence interval (CI)**.  

$\text{Margin of Error} = (Z \text{ or } t) \times \frac{\sigma}{\sqrt{n}}$  
where,  
- $Z$ or $t$  = Critical value from Z-table (large samples) or T-table (small samples)  
- $\sigma$ = Population standard deviation (or $s$ for sample SD)  
- $n$ = Sample size  

**Margin of Error Affect the Confidence Interval :** A confidence interval is calculated as:  
$\text{Confidence Interval} = \bar{x} \pm \text{Margin of Error}$

 - Larger MoE → Wider confidence interval (more uncertainty).  
 - Smaller MoE → Narrower confidence interval (more precise estimate).

##Q16. How is Bayes' Theorem used in statistics, and what is its significance?

ANS 16.

Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update *prior probabilities* based on *new evidence*. It is used to revise beliefs or probabilities when new data becomes available.  

The formula for Bayes' Theorem is:  

$P(A | B) = \frac{P(B | A) \times P(A)}{P(B)}$

where,  
- $P(A | B)$  = probability of event $A$ when $B$ has already occured  
- $P(B | A)$  = probability of event $B$ given $A$ has already occured
- $P(A)$ = independent probability of $A$  
- $P(B)$ =  independent probability of $B$  

**Use of Bayes' Theorem in Statistics (significance) :**  
1. **Medical Diagnosis :** If a patient tests positive for a disease, Bayes' Theorem helps determine the actual probability they have the disease, considering the false positive rate.  

2. **Spam Filtering :**  Email filters use Bayes' Theorem to classify emails as spam or not based on the presence of specific words.  

3. **Machine Learning & Artificial Intelligence :**  Naive Bayes classifiers use Bayes' Theorem to make predictions based on prior data.  

4. **Forensic Science :**  Used in DNA testing and legal cases to calculate the probability of guilt based on forensic evidence.  

5. **Stock Market Predictions :**  Investors update probabilities of market trends based on new financial reports or news.  


##Q17. What is the Chi-square distribution, and when is it used?

ANS 17.

The **Chi-Square ($\chi^2$) distribution** is a probability distribution used in statistical tests that involve categorical data. It describes the distribution of a sum of square of $k$ random variables.
- Chi-square distribution shape is determined by degree of freedom ($k$).  
- It is right-skewed (asymmetrical) but approaches normality as degrees of freedom ($df$) increase.  
- It is a non-negative distribution.

**When is the Chi-Square Distribution Used? :**  
1. **Chi-Square Goodness-of-Fit Test**   
   - Determines if observed categorical data matches an expected distribution.  

2. **Chi-Square Test for Independence :**  
   - Examines whether two categorical variables are independent.  
   - Example: Analyzing whether gender and voting preference are related.  

3. **Chi-Square Test for Homogeneity :**  
   - Compares multiple populations to see if they have the same distribution.  
   - Example: Checking if three brands of cereal have similar popularity.

##Q18. What is the Chi-square goodness of fit test, and how is it applied?

ANS 18.

The **Chi-Square Goodness-of-Fit Test** is a statistical test used to determine whether an observed categorical data distribution *matches an expected distribution*. It helps to assess whether a set of observed frequencies differs significantly from theoretical or expected frequencies.  

- Null Hypothesis ($H_0$): The observed data follows the expected distribution.  
- Alternative Hypothesis ($H_A$): The observed data does not follow the expected distribution.

**Steps to Apply the Chi-Square Goodness-of-Fit Test :**  

Step 1:  Define Hypotheses  
- $H_0$: The observed data follows the expected distribution.  
- $H_A$: The observed data does not follow the expected distribution.  

Step 2: Compute $\chi^2_{Statistics}$
- Use the formula $\chi^2_{Statistics} = \sum \frac{(O - E)^2}{E}$  
where,  
- $O$ = Observed frequency (actual counts in each category)  
- $E$ = Expected frequency (based on theoretical distribution)  
- The sum runs over all categories.  

Step 3: Determine Degrees of Freedom
- $\text{df} = k - 1$, where $k$ is the number of categories.  

Step 4: Find the Critical Value $\chi^2_{Critical}$
- Find $\chi^2_{Critical}$ for $\alpha$ from the *Chi-Square table* at a chosen significance level ( say $\alpha$ = 0.05).
- Compare the calculated $\chi^2_{Statistics}$ value with the $\chi^2_{Critical}$ value.   

Step 5: Make a Decision  
- If $\chi^2_{Statistics}$  $>$  $\chi^2_{Critical}$, then Reject $H_0$ (data does *not* fit expected distribution).  
- If $\chi^2_{Statistics}$  $<$  $\chi^2_{Critical}$, Fail to reject $H_0$ (data *fits* expected distribution).

## Q20. What is an ANOVA test, and what are its assumptions?

ANS 20.

**ANOVA (Analysis of Variance)** is a statistical test used to compare the means of *three or more independent groups* to determine if there is a statistically significant difference between them. Generally, ANOVA is used for more than 2 groups because for 2 groups, we can use T-test.  

- Null Hypothesis ($H_0$): All group means are equal.  
- Alternative Hypothesis ($H_A$): At least one group mean is different.  

ANOVA compares the **variance between groups** to the **variance within groups** using the $F_{Statistics}$. If $F_{Statistics}$ $>$ $F_{Critical}$ , we reject $H_0$, indicating at least one group differs.

**Assumptions of ANOVA :**  
1. Observations within each group must be *independent* (no repeated measures).  

2. The population from which samples are drawn should be normally distributed.

3. **Homogeneity of Variance ;** The variance across groups should be *roughly equal*.  

4. **No Significant Outliers :**  Outliers can distort ANOVA results, so they should be *checked* and *handled*.  


##Q21. What are the different types of ANOVA tests?

ANS 21.

ANOVA (*Analysis of Variance*) is used to compare means across different groups. Depending on the number of factors and the type of data, different types of ANOVA are used.  

**1. One-Way ANOVA (Single-Factor ANOVA) :**  
- Compares the means of three or more independent groups based on one independent variable (factor).
- One factor with atleast two levels, and levels are independent.  
- Example: Comparing *exam scores* of students from *three different schools*.  
- Null Hypothesis ($H_0$): All group means are equal.  
- Alternative Hypothesis ($H_A$): At least one group mean is different.  


**2. Two-Way ANOVA (Factorial ANOVA) :**
- Compares means when **two independent categorical variables (factors)* affect a dependent variable.
- Two or more factors each with atleast two levels, and levels can be either dependent, independent or both.  
- Example: Studying the effect of *diet type (vegetarian, keto, regular)* and *exercise level (low, medium, high)* on *weight loss*.  

**3. Repeated Measures ANOVA :**  
- Compares means when the **same subjects are measured multiple times* under different conditions.
- One factor with atleast two levels but levels are dependent.

- Example: Measuring **memory performance* before, during, and after a cognitive training program.


##Q22. What is the F-test, and how does it relate to hypothesis testing?

ANS 22.

The **F-test** is a statistical test that compares the variances of two or more groups to determine if they are significantly different. It is widely used in hypothesis testing, particularly in *ANOVA (Analysis of Variance)* and *regression analysis*.  

The test is based on the F-distribution, which is always right-skewed and depends on two degrees of freedom ($df1$, $df2$).  

**Formula for the $F_{Statistics}$ :**  
- For comparing variances:  

$F_{Statistics} = \frac{\text{Variance of Sample 1} (S_1^2)}{\text{Variance of Sample 2} (S_2^2)}$


- For ANOVA:  

$F_{Statistics} = \frac{\text{Variance Between Groups}}{\text{Variance Within Groups}}$


A *large F-value* suggests a significant difference between variances or means.  


**Here's how the F-Test relates to Hypothesis Testing :**  

1. *Define Hypotheses*  
   - Null Hypothesis ($H_0$): The variances (or means) are equal.  
   - Alternative Hypothesis ($H_A$): The variances (or means) are not equal.  

2. *Choose Significance Level*  
   - Typically, $\alpha$ = 0.05.  

3. *Compute the $F_{Statistics}$*  
   - Use the appropriate formula based on the test type (e.g., variance test, ANOVA).  

4. *Compare with Critical Value or P-Value*  
   - If $F_{Statistics}$ $>$ $F_{Critical}$ or $p-value$ $<$ 0.05, reject $H_0$ (there is a significant difference).

## Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [52, 55, 49, 50, 53, 51, 54, 48, 50, 52]
population_mean = 50  # Known population mean
population_std = 3  # Known population standard deviation
alpha = 0.05  # Significance level
tail = "two-tailed"

# Sample statistics
sample_size = len(sample)
sample_mean = np.mean(sample)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Determine Z-critical value(s) based on the test type
if tail == "two-tailed":
    z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed test
    rejection_region = f"Z < -{z_critical:.2f} or Z > {z_critical:.2f}"
elif tail == "left-tailed":
    z_critical = norm.ppf(alpha)  # Left-tailed test
    rejection_region = f"Z < {z_critical:.2f}"
else:  # Right-tailed test
    z_critical = norm.ppf(1 - alpha)  # Right-tailed test
    rejection_region = f"Z > {z_critical:.2f}"

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"Z-Critical Value(s): {rejection_region}")

# Decision based on Z-score and critical value
if (tail == "two-tailed" and (z_score < -z_critical or z_score > z_critical)) or \
   (tail == "left-tailed" and z_score < z_critical) or \
   (tail == "right-tailed" and z_score > z_critical):
    print("Conclusion: Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis (H0). There is no significant difference.")
```

##Q2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

```python
import numpy as np
from scipy.stats import norm, t

# Step 1: Define Hypotheses and Simulate Random Data
'''
Null Hypothesis (H0): The sample mean is equal to the hypothesized population mean.
Alternative Hypothesis (H1): The sample mean is significantly different from the hypothesized population mean.
'''
np.random.seed(42)  # For reproducibility
sample_size = 30
population_mean = 50
population_std = 10

sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Define Hypothesis Parameters
hypothesized_mean = 52  # Null hypothesis mean
alpha = 0.05  # Significance level

# Step 3: Compute Sample Statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)

# Step 4: Determine Test Type (Z-test or T-test)
if sample_size >= 30:
    # Use Z-test
    z_score = (sample_mean - hypothesized_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test
    test_type = "Z-test"
else:
    # Use T-test
    t_score = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(sample_size))
    p_value = 2 * (1 - t.cdf(abs(t_score), df=sample_size - 1))  # Two-tailed test
    test_type = "T-test"

# Step 5: Print Results
print(f"Test Type: {test_type}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Test Statistic: {z_score if test_type == 'Z-test' else t_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Step 6: Decision Rule
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis (H0). The sample mean is significantly different from the hypothesized mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis (H0). There is no significant difference.")
```

##Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_0=100, mu_1=105, sigma=15, alpha=0.05, sample_size=30):

    # Compute standard error
    se = sigma / np.sqrt(sample_size)

    # Compute critical value for Type 1 error
    z_critical = norm.ppf(1 - alpha)
    critical_value = mu_0 + z_critical * se  # Right-tailed test boundary

    # Compute Type 2 error probability (Beta)
    beta = norm.cdf((critical_value - mu_1) / se)

    # Generate x values for plotting
    x = np.linspace(mu_0 - 4 * sigma, mu_1 + 4 * sigma, 1000)

    # Null and alternative distributions
    y_null = norm.pdf(x, mu_0, se)  # Under H0
    y_alt = norm.pdf(x, mu_1, se)   # Under H1

    # Plot distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, y_null, label="Null Distribution (H0)", color='blue')
    plt.plot(x, y_alt, label="Alternative Distribution (H1)", color='green')

    # Shade Type 1 Error (α)
    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.5, label="Type 1 Error (α)")

    # Shade Type 2 Error (β)
    plt.fill_between(x, y_alt, where=(x <= critical_value), color='purple', alpha=0.5, label="Type 2 Error (β)")

    # Plot critical value line
    plt.axvline(critical_value, color='black', linestyle='dashed', label=f"Critical Value: {critical_value:.2f}")

    # Labels and title
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()

    # Show plot
    plt.show()

    # Print calculated values
    print(f"\nCritical Value: {critical_value:.2f}")
    print(f"Type 1 Error (α): {alpha:.2f}")
    print(f"Type 2 Error (β): {beta:.2f}")
    print(f"Power of the Test (1 - β): {1 - beta:.2f}")

# Example Usage
visualize_errors()
```

##Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95, pop_std=None):

    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation
    se = sample_std / np.sqrt(n)  # Standard error

    if pop_std:  # Use Z-distribution if population std is known
        critical_value = stats.norm.ppf(1 - (1 - confidence) / 2)
    else:  # Use T-distribution if population std is unknown
        critical_value = stats.t.ppf(1 - (1 - confidence) / 2, df=n-1)

    margin_of_error = critical_value * se
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample

# Compute CI using T-distribution (since population std is unknown)
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

**What Does This Confidence Interval Mean?**   
The 95% confidence interval (CI) tells us that we are 95% confident that the true population mean falls between 92.14 and 102.22.

**Significance:**  
1. Helps in Decision Making  
If a claimed population mean (e.g., μ = 100) is inside the interval → No strong evidence against it.
If a claimed population mean is outside the interval → Likely false.
2. Gives Precision of the Estimate  
- Narrow CI → More precise estimate (higher confidence in the sample mean).
- Wide CI → More variability (less confidence, usually due to small sample size).


3. Alternative to Hypothesis Testing  
If the CI excludes 100, it suggests a significant difference.
If 100 is within the interval, we cannot say the sample mean is significantly different.

**Example Interpretation in Context :**  
Let's say a pharmaceutical company claims that the average effectiveness of a drug is 100 mg/L.

Our 95% CI = (92.14, 102.22) → The true mean could be 100, so we fail to reject their claim.
If our CI was (107, 115), we could say their claim is likely false (because 100 is outside the range).

##Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive):
    not_h = 1 - prior  # P(-H): Probability of NOT having the disease
    evidence = (likelihood * prior) + (false_positive * not_h)  # P(E)

    posterior = (likelihood * prior) / evidence  # P(H|E)
    return posterior

# Given probabilities
prior = 0.01       # P(H): 1% disease prevalence
likelihood = 0.95  # P(E|H): Sensitivity (true positive rate)
false_positive = 0.05  # P(E|-H): False positive rate

posterior = bayes_theorem(prior, likelihood, false_positive)
print(f"Posterior Probability (P(H|E)): {posterior:.4f}")
```

**Explanation of the Process :**
1. Define prior probability
$𝑃(𝐻)$ → The initial belief (e.g., 1% population has the disease).
2. Define likelihood
$𝑃(𝐸∣𝐻)$ → The probability of a positive test given the person has the disease.
3. Define false positive rate
$𝑃(𝐸|-𝐻)$ → The probability of a positive test even if the person does NOT have the disease.
4. Compute total evidence probability
$𝑃(𝐸)$ → The probability that any random person tests positive, regardless of their actual disease status.
5. Compute posterior probability
$𝑃(𝐻∣𝐸)$ → Update our belief based on new evidence.



This means that even after testing positive, the probability of actually having the disease is only ~16.1%, due to the high false-positive rate.

##Q12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import scipy.stats as stats

#Hypothesis:
'''
H0: The two categorical variables are independent (no association).
HA: The two categorical variables are dependent (there is an association).
'''

# Observed Frequencies (O)
observed = np.array([[30, 10],   # Males: 30 prefer A, 10 prefer B
                      [20, 40]])  # Females: 20 prefer A, 40 prefer B

# Compute row and column totals
row_totals = observed.sum(axis=1).reshape(-1, 1)
col_totals = observed.sum(axis=0)
total = observed.sum()

# Compute Expected Frequencies (E) using (Row Total * Column Total) / Grand Total
expected = (row_totals @ col_totals.reshape(1, -1)) / total

# Compute Chi-square statistic
chi_square_stat = np.sum((observed - expected) ** 2 / expected)

# Degrees of Freedom
dof = 2-1

# Compute P-value
p_value = 1 - stats.chi2.cdf(chi_square_stat, dof)

# Display results
print("Chi-Square Test for Independence")
print("----------------------------------")
print("Hypotheses:")
print("H0: The two categorical variables are independent (no association).")
print("HA: The two categorical variables are dependent (there is an association).")
print("\nResults:")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:\n{expected}")

# Decision
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nDecision: Reject the Null Hypothesis (H0). The variables are dependent.")
else:
    print("\nDecision: Fail to Reject the Null Hypothesis (H0). The variables are independent.")
```

##Q16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Define Hypotheses random seed for reproducibility
'''
H0: The variances of the two populations are equal (σ1² = σ2²).
HA: The variances of the two populations are NOT equal (σ1² ≠ σ2²).
'''

np.random.seed(42)

# Step 2: Generate two random samples
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, n=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15, n=30

# Step 3: Compute sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 4: Compute F-Statistic
# Formula: F = Larger Variance / Smaller Variance
F_statistic = var1 / var2

# Step 5: Compute Degrees of Freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Step 6: Compute P-Value (Two-Tailed Test)
p_value = 2 * (1 - stats.f.cdf(F_statistic, df1, df2))

# Step 7: Print results
print("F-Test for Equal Variances")
print("--------------------------")

print("\nStep 1: Define Hypotheses")
print("H0: The variances of the two populations are equal (σ1² = σ2²).")
print("HA: The variances of the two populations are NOT equal (σ1² ≠ σ2²).")

print("\nStep 2: Compute Sample Variances")
print(f"Sample Variance 1: {var1:.4f}")
print(f"Sample Variance 2: {var2:.4f}")

print("\nStep 3: Compute F-Statistic")
print(f"F-Statistic: {F_statistic:.4f}")

print("\nStep 4: Compute P-Value")
print(f"P-Value: {p_value:.4f}")

print("\nStep 5: Compare with Significance Level (α = 0.05)")

# Step 8: Decision
alpha = 0.05

if p_value < alpha:
    print("\nDecision: Reject the Null Hypothesis (H0). Variances are significantly different.")
else:
    print("\nDecision: Fail to Reject the Null Hypothesis (H0). Variances are not significantly different.")
```

##Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.stattools import durbin_watson

def check_anova_assumptions(*groups):
    print("\n Checking Assumptions for ANOVA")
    print("====================================")

    # 1️ Normality Test (Shapiro-Wilk Test)
    print("\n Normality Test :")
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: W = {stat:.4f}, p = {p_value:.4f} {' Normal' if p_value > 0.05 else ' Not Normal'}")

    # 2️ Equal Variance Test
    print("\n Homogeneity of Variance Test :")
    stat, p_value = stats.levene(*groups)
    print(f"W = {stat:.4f}, p = {p_value:.4f} {' Variances are Equal' if p_value > 0.05 else ' Variances are Not Equal'}")

    # 3️ Independence Assumption
    print("\n Independence Check (Durbin-Watson Test)")
    combined_data = np.concatenate(groups)  # Merge all groups into one dataset
    dw_statistic = durbin_watson(combined_data)

    if 1.5 < dw_statistic < 2.5:
        independence_result = "Data appears independent"
    else:
        independence_result = "Possible autocorrelation (Not independent)"
    print(f"Durbin-Watson Statistic: {dw_statistic:.4f} → {independence_result}")

# Example Usage
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Call the function
check_anova_assumptions(group1, group2, group3)
```

##Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

'''Hypotheses:

Main Effect of Exercise Type:
H0: No significant effect of exercise type.
HA: Exercise type affects weight loss.

Main Effect of Diet Type:
H0: No significant effect of diet type.
HA: Diet type affects weight loss.

Interaction Effect (Exercise × Diet):
H0: No interaction between exercise and diet.
HA:Exercise and diet interact to affect weight loss.
'''

# Simulate Data
np.random.seed(42)
data = {
    "Exercise": np.repeat(["Cardio", "Strength"], 30),
    "Diet": np.tile(["Low-Carb", "High-Protein", "Balanced"], 20),
    "Weight_Loss": np.random.normal(loc=5, scale=1.5, size=60) +
                   (np.repeat([1, -1], 30)) + (np.tile([1, -0.5, 0], 20))
}
df = pd.DataFrame(data)

#Fit Two-Way ANOVA Model
model = ols("Weight_Loss ~ C(Exercise) + C(Diet) + C(Exercise):C(Diet)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nTwo-Way ANOVA Results:\n")
print(anova_table)

#Visualizing Results
plt.figure(figsize=(8, 5))
sns.boxplot(x="Diet", y="Weight_Loss", hue="Exercise", data=df, palette="Set2")
plt.title("Effect of Exercise & Diet on Weight Loss")
plt.xlabel("Diet Type")
plt.ylabel("Weight Loss (kg)")
plt.legend(title="Exercise Type")
plt.grid()
plt.show()
```

##Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the result.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Hypotheses
'''
H0: The population variance is equal to a specified value
HA: The population variance is different from specified value

'''
#Simulate Data (Normal Distribution)
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30
sample_size = len(sample)
sample_variance = np.var(sample, ddof=1)  # Sample variance (unbiased)

#Hypothesized Population Variance
sigma_squared_0 = 100

#Compute Chi-Square Statistic
chi_square_stat = (sample_size - 1) * sample_variance / sigma_squared_0

#Critical Values for Two-Tailed Test (α = 0.05)
alpha = 0.05
df = sample_size - 1  # Degrees of freedom
chi_critical_low = stats.chi2.ppf(alpha / 2, df)
chi_critical_high = stats.chi2.ppf(1 - alpha / 2, df)

#Print Results
print("\nHypothesis Testing for Population Variance")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"Critical Values: ({chi_critical_low:.4f}, {chi_critical_high:.4f})")

#Interpretation
if chi_square_stat < chi_critical_low or chi_square_stat > chi_critical_high:
    print("\n Conclusion: Reject the null hypothesis (H0). The population variance is significantly different.")
else:
    print("\n Conclusion: Fail to reject the null hypothesis (H0). The population variance is not significantly different.")

#Visualizing the Chi-Square Distribution
x = np.linspace(chi_critical_low * 0.8, chi_critical_high * 1.2, 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})', color='blue')
plt.axvline(chi_square_stat, color='black', linestyle="dashed", label=f'Chi-Square Statistic')
plt.axvline(chi_critical_low, color='red', linestyle="dashed", label="Lower Critical Value")
plt.axvline(chi_critical_high, color='red', linestyle="dashed", label="Upper Critical Value")

plt.fill_between(x, y, 0, where=(x < chi_critical_low) | (x > chi_critical_high), color='red', alpha=0.3, label="Rejection Regions")

# Labels & Legend
plt.title("Chi-Square Test for Variance")
plt.xlabel("Chi-Square Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

##Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

#Hypotheses
'''
H0: The proportions in both groups are equal.
HA: The proportions in both groups are different.
'''

#Given Data (Successes & Sample Sizes for Two Groups)
success_1 = 45  # Number of successes in group 1
size_1 = 100    # Sample size of group 1

success_2 = 30  # Number of successes in group 2
size_2 = 100    # Sample size of group 2

#Compute Sample Proportions
p1 = success_1 / size_1  # Proportion for group 1
p2 = success_2 / size_2  # Proportion for group 2

#Compute Pooled Proportion
p_pool = (success_1 + success_2) / (size_1 + size_2)

#Compute Standard Error (SE)
se = np.sqrt(p_pool * (1 - p_pool) * (1 / size_1 + 1 / size_2))

#Compute Z-Statistic
z_stat = (p1 - p2) / se

#Compute Critical Value for Two-Tailed Test (α = 0.05)
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

#Print Results
print("\nZ-Test for Comparing Proportions")
print(f"Proportion 1: {p1:.4f}, Proportion 2: {p2:.4f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"Critical Value (±{z_critical:.4f})")

#Decision
if abs(z_stat) > z_critical:
    print("\nConclusion: Reject the null hypothesis (H0). The proportions are significantly different.")
else:
    print("\nConclusion: Fail to reject the null hypothesis (H0). No significant difference in proportions.")
```

##Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Hypotheses
'''
H0: The variances of the two groups are equal.
HA: The variances of the two groups are significantly different.
'''

#Simulate Data for Two Groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=15, size=30)

#Compute Sample Variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

#Compute F-Statistic (Always > 1 for consistency)
F_stat = var1 / var2 if var1 > var2 else var2 / var1  # Max variance / Min variance

#Find Critical Value (α/2 for two-tailed test)
alpha = 0.05
df1, df2 = len(group1) - 1, len(group2) - 1
F_critical = stats.f.ppf(1 - alpha / 2, df1, df2)

#Print Results
print("\nF-Test for Variance Comparison (Two-Tailed)")
print(f"Variance 1: {var1:.4f}, Variance 2: {var2:.4f}")
print(f"F-Statistic: {F_stat:.4f}")
print(f"F-Critical Value: {F_critical:.4f}")

#Interpretation
if F_stat > F_critical:
    print("\nConclusion: Reject the null hypothesis (H0). The variances are significantly different.\n")
else:
    print("\nConclusion: Fail to reject the null hypothesis (H0). The variances are not significantly different.\n")

#Visualizing the F-Distribution
x = np.linspace(0, F_critical * 1.2, 1000)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'F-Distribution (df1={df1}, df2={df2})', color='blue')
plt.axvline(F_stat, color='black', linestyle="dashed", label=f'F-Statistic')
plt.axvline(F_critical, color='red', linestyle="dashed", label="F-Critical Value")

plt.fill_between(x, y, 0, where=(x > F_critical), color='red', alpha=0.3, label="Rejection Region")

plt.title("F-Test for Variance Comparison (Two-Tailed)")
plt.xlabel("F-Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

**Interpretation :**
- If $F_{stat} > F_{critical}$, Reject $H_0$ → Variances are significantly different.
- If $F_{stat} \leq F_{critical}$, Fail to reject $H_0$ → Variances are not significantly different.

##Q27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

**Scenario: Distribution of Customers' Favorite Coffee Types**  
A coffee shop surveyed 200 customers about their favorite type of coffee. The shop expects the preferences to be uniformly distributed among four types: Espresso, Latte, Cappuccino, and Americano. However, the actual survey data may differ. We will use a Chi-Square Goodness of Fit Test to check if customer preferences match the expected distribution.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Hypotheses
'''
H0: The observed customer preferences follow the expected uniform distribution.
HA: The observed customer preferences do not follow the expected uniform distribution.
'''

#Observed customer preferences (from the survey)
observed = np.array([59, 96, 49, 75])  # Espresso, Latte, Cappuccino, Americano

#Expected frequencies (assuming uniform preference among 4 types)
total_customers = np.sum(observed)
expected = np.full(len(observed), total_customers / len(observed))

#Perform Chi-Square Goodness of Fit Test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
df = len(observed) - 1  # Degrees of freedom (df = categories - 1)
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)  # Critical value from Chi-Square table

#Print Results
print("\nChi-Square Goodness of Fit Test (Customer Coffee Preferences)")
print(f"Observed Preferences: {observed}")
print(f"Expected Preferences: {expected}")
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical Value (α=0.05): {critical_value:.4f}")
print(f"P-Value: {p_value:.4f}")

#Interpretation
if chi2_stat > critical_value:
    print("\nConclusion: Reject the null hypothesis (H0). The observed coffee preferences do NOT match the expected uniform distribution.\n")
else:
    print("\nConclusion: Fail to reject the null hypothesis (H0). The observed coffee preferences match the expected uniform distribution.\n")

#Visualizing the Chi-Square Distribution
x = np.linspace(0, chi2_stat * 1.5, 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})', color='blue')
plt.axvline(chi2_stat, color='black', linestyle="dashed", label=f'Chi-Square Statistic')
plt.axvline(critical_value, color='red', linestyle="dashed", label="Critical Value (α=0.05)")

plt.fill_between(x, y, 0, where=(x > critical_value), color='red', alpha=0.3, label="Rejection Region")

plt.title("Chi-Square Distribution for Coffee Preference Test")
plt.xlabel("Chi-Square Value")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()

#Bar Plot: Observed vs. Expected Preferences
categories = ["Espresso", "Latte", "Cappuccino", "Americano"]
bar_width = 0.4
plt.figure(figsize=(8, 5))
plt.bar(categories, observed, width=bar_width, label="Observed", color="blue", alpha=0.7)
plt.bar(categories, expected, width=bar_width, label="Expected", color="red", alpha=0.7, edgecolor="black")

plt.xlabel("Coffee Type")
plt.ylabel("Number of Customers")
plt.title("Observed vs. Expected Coffee Preferences")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

**Interpretation :**
Since **chi2_stat > critical_value** $($i.e., $\chi^2_{Statistics} > \chi^2_{Critical})$, we reject $H_0$ → Customer preferences are not evenly distributed. Some coffee types are more popular than expected.


**Real-World Significance :**
If the coffee shop rejects $H_0$, it means certain coffee types are more popular than others, so they can adjust inventory and marketing strategies accordingly.  
If the test fails to reject $H_0$, customer preferences match the expectation, so no major changes are needed.
This test helps businesses understand customer behavior using real data.
